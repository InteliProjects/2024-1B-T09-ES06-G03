## 2
 Verificando a distribuição da quantidade de avaliações por projeto

```python
df3.id_projeto.value_counts()
```

```python
contagem = df3['id_projeto'].value_counts()

# Plotar o gráfico de barras
plt.figure(figsize=(30, 10))
contagem.plot(kind='bar')
plt.title('Quantidade de avaliações por projeto')
plt.xlabel('Id do projeto')
plt.ylabel('Quantidade de avaliações')
plt.show()
```
